Loading The Libraries and Dataset

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")

df=pd.read_csv("Downloads/IPL.csv")
df.head()
 
```

1.which team won the most matches?

```python
match_winner=df["match_winner"].value_counts()
match_winner
```

```python
sns.barplot(y=match_winner.index,x=match_winner.values,palette="viridis")
plt.title("Most match win by team")
```

2.Toss decision trends

```python
sns.countplot(x=df["toss_decision"],palette="rainbow")
plt.title("Toss decision trends")
```

3.Toss Winner vs Match Winner

```python
count=df[df["toss_winner"]==df["match_winner"]]["match_id"].count()
percentage=(count*100)/df.shape[0]
percentage.round(2)
```

4.How do teams win? (Runs vs Wickets)


```python
sns.countplot(x=df["won_by"],palette="viridis")
plt.title("WON_BY")
```

Key player performance

1.Most "player of the match" Awards

```python
count1=df["player_of_the_match"].value_counts().head(10)
count1
```

```python
sns.barplot(y=count1.index,x=count1.values,palette="mako")
plt.title("Top 10 Player's of the match")
```

2 top scores

```python
high=df.groupby("top_scorer")["highscore"].sum().sort_values(ascending=False).head(2)
high
```

```python
high.plot(kind="bar")
plt.ylabel("highscore")
plt.title("Top 2 score")
```

10 best bowling figures

```python
df.head()
```

```python
df["highest_wickets"]=df["best_bowling_figure"].apply(lambda x:x.split("--")[0])
df["highest_wickets"]=df["highest_wickets"].astype(int)
top_bowlers=df.groupby("best_bowling")["highest_wickets"].sum().sort_values(ascending=False).head(10)
top_bowlers.plot(kind="barh")
```
